```python
kalimat_1 = "Saya suka belajar pemrograman Python"
kalimat_2 = "Python adalah bahasa pemrograman yang populer"
kalimat_3 = "Saya belajar Python di VSCode menggunakan WSL"
```

Ekstraksi kata-kata unik dari ketiga kalimat

```python
tokens = []
for word in kalimat_1.lower().split():
    if word not in tokens:
        tokens.append(word)
        
for word in kalimat_2.lower().split():
    if word not in tokens:
        tokens.append(word)

for word in kalimat_3.lower().split():
    if word not in tokens:
        tokens.append(word)
```

```python
tokens
```

Membuat representasi vektor untuk ketiga kalimat

```python
vector_one = []
for token in tokens:
    if token in kalimat_1.lower().split():
            vector_one.append(1)
    else:
            vector_one.append(0)
```

```python
vector_one
```

```python
vector_two = []
for token in tokens:
    if token in kalimat_2.lower().split():
            vector_two.append(1)
    else:
            vector_two.append(0)
```

```python
vector_three = []
for token in tokens:
    if token in kalimat_3.lower().split():
            vector_three.append(1)
    else:
            vector_three.append(0)
```

```python
vector_three
```

Kalimat mana yang mengandung kata 'pemrograman' tetapi tidak mengandung kata 'VSCode'?

```python
pemrograman_vector = [1,1,0]
print(f'Vector for pemrograman : {pemrograman_vector}')
```

```python
vscode_vector = [0,0,1]
print(f'Vector for vscode : {vscode_vector}')
```

```python
not_vscode_vector = [1,1,0]
```

```python
result_vector = []
for i in range(len(pemrograman_vector)):
    # For 1 & 1
    if (pemrograman_vector[i] == 1 and not_vscode_vector[i] == 1):
        result_vector.append(1)
    # For 1 & 0 or 0 & 1
    elif (pemrograman_vector[i] == 1 and not_vscode_vector[i] == 0) or (pemrograman_vector[i] == 0 and not_vscode_vector[i] == 1):
        result_vector.append(0)
    else:
        result_vector.append(0)
print(f'Final vector post BITWISE AND operator : {result_vector}')
```

Oleh karena itu, kita menyimpulkan bahwa kalimat satu dan kalimat dua mengandung istilah 'cow' tetapi tidak mengandung 'tuesday'.

Indeks Terbalik

```python
sorted_tokens = sorted(tokens)
```

```python
sorted_tokens
```

```python
token_frequency = []
for token in sorted_tokens:
    token_frq = 0
    for word in kalimat_1.lower().split():
        if word == token:
            token_frq += 1
    for word in kalimat_2.lower().split():
        if word == token:
            token_frq += 1
    for word in kalimat_3.lower().split():
        if word == token:
            token_frq += 1
    token_frequency.append(token_frq)
            
```

```python
token_frequency
```

```python
posting_list = []
for token in sorted_tokens:
    p_list = ""
    for word in kalimat_1.lower().split():
        if word == token and  "1 -->" not in p_list:
            p_list += "1 -->"
    for word in kalimat_2.lower().split():
        if word == token and  "2 -->" not in p_list:
            p_list += "2 -->"
    for word in kalimat_3.lower().split():
        if word == token and  "3 -->" not in p_list:
            p_list += "3 -->"
    posting_list.append(p_list)         
```

```python
posting_list
```

```python
posting_list_df = pd.DataFrame(columns = ['words','term_frequency','posting_list'])
```

```python
posting_list_df['words'] = sorted_tokens
posting_list_df['term_frequency'] = token_frequency
posting_list_df['posting_list'] = posting_list
```

```python
print(kalimat_1 + '\n' + kalimat_2 + '\n' + kalimat_3)
posting_list_df
```

```python
input_search = input("Enter the search string : ")
```

```python
input_search
```

```python
filter_from_posting_list = []
for token in input_search.split():
    filter_from_posting_list.append(posting_list_df[posting_list_df.words == token]['posting_list'].item())
```

```python
filter_from_posting_list
```

```python
searched_list = []
for x in filter_from_posting_list:
    for y in x.split(' -->'):
        if y is not '':
            searched_list.append(y)
```

```python
print(f'Documents which are retrived based on the input strings are - {set(sorted(searched_list))}')
```
